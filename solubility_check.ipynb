# Importing Dataset


```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
df
```

# Data Preparation

## Seperation of data in x and y

```python
y = df['logS']
y
```

```python
X = df.drop('logS', axis=1)
X
```

## Seperation into training and testing data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state=100)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

# Model Building

## Linear Regression

### **Training the model**

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

###**Applying the model to make predictions**

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

### **Evaluating Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
lr_results = pd.DataFrame(['LinearRegression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method of Evaluation', 'Training MSE', 'Training R2', 'Testing MSE', 'Training R2']
lr_results
```

## Random Forest

###**Training the model**

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth = 2, random_state=100)
rf.fit(X_train, y_train)
```

###**Applying the model to make predictions**

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

### **Evaluating Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['RandomForest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method of Evaluation', 'Training MSE', 'Training R2', 'Testing MSE', 'Training R2']
rf_results
```

##Model Comparision

```python
df_models = pd.concat([lr_results, rf_results], axis=0).reset_index(drop=True)
df_models
```
